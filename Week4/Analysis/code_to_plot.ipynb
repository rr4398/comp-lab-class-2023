```python
import mdtraj as md
import matplotlib.pyplot as plt

traj = md.load('trajectory.xtc', top='topology.pdb')
# Calculate the end-to-end distance
end_to_end_distances = md.compute_distances(traj, [[0, traj.n_atoms - 1]])

# Calculate the radius of gyration
radius_of_gyration = md.compute_rg(traj)
time = traj.time
# Create a new figure and two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 6))

# Plot end-to-end distance vs. time
ax1.plot(time, end_to_end_distances, label='End-to-End Distance', color='green')
ax1.set_ylabel('Distance (nm)')
ax1.legend(loc='upper right')

# Plot radius of gyration vs. time
ax2.plot(time, radius_of_gyration, label='Radius of Gyration',color='red')
ax2.set_xlabel('Time (ps)')
ax2.set_ylabel('Radius of Gyration (nm)')
ax2.legend(loc='upper right')

# Save the figures to the 'Figures' directory
fig.savefig('distance_and_gyration.pdf')

# Show the plots in the Jupyter notebook
plt.show()

```
