```python
import matplotlib.pyplot as plt

# Function to read data from a .txt file
def read_data_from_file(filename):
    x_data = []
    y_data = []
    with open(filename, 'r') as file:
        lines = file.readlines()
    for line in lines:
        parts = line.split()
        if len(parts) == 2:
            x, y = map(float, parts)
            x_data.append(x)
            y_data.append(y)
    return x_data, y_data

# Read data from the first .txt file
x_data1, y_data1 = read_data_from_file('rmsd.txt')

# Read data from the second .txt file
x_data2, y_data2 = read_data_from_file('rmsd_xtal.txt')

# Create a plot and specify different colors
plt.plot(x_data1, y_data1, label='Equilibrated', color='blue')
plt.plot(x_data2, y_data2, label='Crystal', color='red')

# Add labels and a legend
plt.xlabel('Time(ns)')
plt.ylabel('RMSD(nm)')
plt.title('RMSD: 1AKI, Backbone')
plt.legend()

# Save the plot as a PDF
plt.savefig('plot_rmsd.pdf', format='pdf')

# Show the plot (optional)
plt.show()
```
