```python
import mdtraj as md
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf as pdf
import numpy as np
```

```python
traj=md.load_trr('../Data/T300/adp_md.trr',top='../Data/adp.gro')
```

```python
phi_indices, phi_angles=md.compute_phi(traj)
psi_indices, psi_angles=md.compute_psi(traj)
```

```python
plt.plot(traj.time, phi_angles, label='Phi', alpha=0.7)

plt.title('Phi vs Time')
plt.xlabel('Time (ps)')
plt.ylabel('Dihedral Angle (radians)')
plt.legend(bbox_to_anchor=(1,0.6))

pdf_file=pdf.PdfPages('../Figures/Week5_figures.pdf')
pdf_file.savefig()
```

```python
plt.plot(traj.time, psi_angles, label='Psi', alpha=0.7)
plt.title('Psi vs Time')
plt.xlabel('Time (ps)')
plt.ylabel('Dihedral Angle (radians)')
plt.legend(bbox_to_anchor=(1,0.6))

pdf_file.savefig()
```

```python
# 2D histogram
histogram, xedges, yedges = np.histogram2d(phi_angles.ravel(), psi_angles.ravel(), bins=50)

# Avoid log(0) by adding a small number
histogram += 1e-7

# Compute the free energy surface
F = -np.log(histogram)

# Subtract the minimum free energy
F -= F.min()

# Plotting the free energy surface

plt.contourf(xedges[:-1], yedges[:-1], F.T, 40, cmap='viridis')
cbar = plt.colorbar(label='Free energy (kT)')
plt.xlabel('Phi (radians)')
plt.ylabel('Psi (radians)')
plt.title('Free Energy Surface in Phi-Psi Space')

pdf_file.savefig()
```

```python
def compute_histogram(energy_data, bins=100):
    hist, edges = np.histogram(energy_data, bins=bins, density=True)
    bin_centers = (edges[:-1] + edges[1:]) / 2.0
    return bin_centers, hist


T300_pote = np.loadtxt('../Data/potential_T300.xvg', comments=['@', '#'])[:,1]
T363_pote = np.loadtxt('../Data/potential_T363.xvg', comments=['@', '#'])[:,1]
T440_pote = np.loadtxt('../Data/potential_T440.xvg', comments=['@', '#'])[:,1]
```

```python
T300_bin, hist = compute_histogram(T300_pote)
plt.plot(T300_bin, hist, label="T300")

T363_bin, hist = compute_histogram(T363_pote)
plt.plot(T363_bin, hist, label="T363")

T440_bin, hist = compute_histogram(T440_pote)
plt.plot(T440_bin, hist, label="T440")


plt.xlabel('Potential Energy')
plt.ylabel('Density')
plt.legend()
plt.title('Potential Energy Histograms of Replicas')
plt.tight_layout()

pdf_file.savefig()
```

```python
k_B = 8.617333262145e-5  # Boltzmann constant in eV/K
T = 300  # Temperature in K

# Add a small value to avoid taking the log of zero
histogram += 1e-10

free_energy = -k_B * T * np.log(histogram)
free_energy -= free_energy.min()  # Shift so that the minimum value is 0

plt.contourf((xedges[:-1] + xedges[1:]) / 2, (yedges[:-1] + yedges[1:]) / 2, free_energy.T, levels=20)
plt.colorbar(label='Free Energy (eV)')
plt.xlabel('Phi')
plt.ylabel('Psi')
plt.title('Free Energy Surface at T=300K')

pdf_file.savefig()

```

```python
time,*temperatures=np.loadtxt('../replica_index.xvg',unpack=True)

# Considering every 100th frame
time = time[::100]
temperatures = [temp[::100] for temp in temperatures]

# Plotting each replica
for idx, temp in enumerate(temperatures, start=1):
    plt.plot(time, temp, label=f'Replica {idx}')

plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Temperature Movement of Each Replica')
plt.legend()
plt.tight_layout()

pdf_file.savefig()
pdf_file.close()
```
